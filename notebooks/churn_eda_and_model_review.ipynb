# Customer Retention Decision System

This notebook walks through the data audit, churn patterns, model comparison, and retention framing used in the project.

```python
import pandas as pd
from IPython.display import display

from src.data.loader import load_clean_data

df = load_clean_data()
df.head()
```

```python
summary = {
    "rows": len(df),
    "columns": df.shape[1],
    "churn_rate": round(df['Churn'].mean(), 4),
    "blank_total_charges_fixed": int((df['TotalCharges'] == 0).sum()),
}
summary
```

```python
contract_churn = (df.groupby('Contract')['Churn'].mean() * 100).sort_values(ascending=False).round(2)
tech_support_churn = (df.groupby('TechSupport')['Churn'].mean() * 100).sort_values(ascending=False).round(2)
payment_churn = (df.groupby('PaymentMethod')['Churn'].mean() * 100).sort_values(ascending=False).round(2)

display(contract_churn.rename('churn_rate_pct').to_frame())
display(tech_support_churn.rename('churn_rate_pct').to_frame())
display(payment_churn.rename('churn_rate_pct').to_frame())
```

```python
validation_metrics = pd.read_csv('../models/artifacts/validation_model_metrics.csv')
test_metrics = pd.read_csv('../models/artifacts/test_metrics.csv')
feature_importance = pd.read_csv('../models/artifacts/feature_importance.csv').head(10)

display(validation_metrics)
display(test_metrics)
display(feature_importance)
```

## Takeaways

- Month-to-month customers churn far more often than customers on longer contracts.
- Lack of tech support and electronic check payments are strong churn signals.
- Logistic regression was selected because it delivered the strongest validation PR-AUC while staying easy to interpret.
- A 0.60 threshold improved precision without giving up too much recall, which is helpful for targeted retention spend.
